# Matplotlib 基础知识学习


```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

%matplotlib inline
rcParams['figure.figsize'] = 10,8
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-5,5,0.1)           # [-5,5) with step 0.1
y = x * x

plt.figure(1)
plt.title("y=x^2")
plt.plot(x,y)
plt.show()
```
